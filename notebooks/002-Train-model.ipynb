```
from pickle import load
from mlxtend.utils.data import *
X_train = load(open(str(get_file_path(name="X", extension=".pkl")), 'rb'))
y_train = load(open(str(get_file_path(name="y", extension=".pkl")), 'rb'))

# X_train, y_train = random_extract(X_train, y_train, seed=0, ratio=0.1) # uncomment to use a smaller train set and make model training faster
```

```
from autokeras import ImageClassifier
model = ImageClassifier(max_trials=1)
model.fit(X_train, y_train, epochs=1)
```

```
model.export_model().save(str(get_data_path().joinpath("model")), save_format="tf")
```
