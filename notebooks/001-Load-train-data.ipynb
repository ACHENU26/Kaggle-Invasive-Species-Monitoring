## Outputs

The outputs (labels) are in a CSV file:

```python
from pandas import read_csv
from mlxtend.utils.data import *

data_path = get_data_path()
labels = read_csv(str(data_path.joinpath("train_labels.csv")))
labels.head(5)
```

## Inputs

Get list of image file names and shuffle:

```python
from os import listdir
from os.path import isfile, join
import random

train_data_path = data_path.joinpath("train")
images = [f for f in listdir(train_data_path) if isfile(train_data_path.joinpath(f))]
random.shuffle(images)
images = images[0:50]
```

Define and apply a function to load images and downsize them (to make model training easier):

```python
from tensorflow.keras.preprocessing import image
from tqdm import tqdm # to display progress bar when running loops

X_train = []
y_train = []

smallimg_size = 224
for f in tqdm(images, miniters=10):
    im_number = int(f.split('.jpg')[0])
    y_train.append(int(labels[labels['name']==im_number]['invasive']))
    img = image.load_img(train_data_path.joinpath(f), target_size=(smallimg_size, smallimg_size))
    X_train.append(image.img_to_array(img))
```

The result of this is 2 lists, but we'll need arrays in order to train a model.

```python
import numpy as np
X_train = np.array(X_train, np.float32)
y_train = np.array(y_train)
```

## Save data

```python
from pickle import dump
dump(X_train, open(str(get_file_path(name="X_train", extension=".pkl")), 'wb'))
dump(y_train, open(str(get_file_path(name="y_train", extension=".pkl")), 'wb'))
```
